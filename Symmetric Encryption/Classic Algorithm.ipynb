# vigenère :

It’s just an improved Caesar cipher , in this cipher the key can have more than one letter .

the key in the example below has 3 digits CODEX .




```python
def encrypt(message, key):
    result = ""
    i=0
    for char in message:
        if (char.isalpha()):
            
            #uppercase
            if (char.isupper()):
                result += chr( ( ord(char) + ord(key[i%len(key)].upper()) - 2*65) % 26 + 65  )
            #lowercase
            else :
                result += chr((ord(char) + ord(key[i%len(key)].lower()) - 2*97) % 26 +97)    
            i += 1
        else :
            result += " "
            
    return result
```

```python
def decrypt(message, key):
    result = ""
    i=0
    for char in message:
        if (char.isalpha()):
            #uppercase
            if (char.isupper()):
                result += chr( ( ord(char) - ord(key[i%len(key)].upper())) % 26 + 65  )
            #lowercase
            else :
                result += chr((ord(char) - ord(key[i%len(key)].lower()) ) % 26 + 97) 
            i += 1
        else :
            result += " "
            
    return result
```

```python
message = "i love garage"
key = "iAmLaZa"
encrypted = encrypt(message, key)
decrypted = decrypt(encrypted, key)

print("Original message: " + message)
print("Encrypted message: " + encrypted)
print("Decrypted message: " + decrypted)
```

# Affine :
The affine cipher is a cryptography method based on a mono-alphabetic substitution cipher, that is to say that the original letter is only replaced by a single other letter.



Encryption :E = (a*x +b) % [26]
Decryption :D = (a^(-1)*E -b) % [26]



```python
from Crypto.Util.number import inverse
```

```python
def encrypt_affine(text, a , b):
    result= ""
    #plaintext -> ciphertext
    for char in text:
        if(char.isalpha()):
            #uppercase
            if (char.isupper()):
                result += chr(( a*(ord(char) - 65) +b ) % 26 + 65 )
            #lowercase
            else :
                result += chr(( a*(ord(char) - 97) +b ) % 26 + 97 )
        else :
            result += " "
    return result 
```

```python
def decrypt_affine(text, a , b):
    result= ""
    #plaintext -> ciphertext
    for char in text:
        if(char.isalpha()):
            #uppercase
            if (char.isupper()):
                result += chr(( a*(ord(char) - 65) -b ) % 26 + 65 )
            #lowercase
            else :
                result += chr(( a*(ord(char) - 97) -b ) % 26 + 97 )
        else :
            result += " "
    return result 
```

```python
message = "I lovE Garage"
encrypted = encrypt_affine(message, 1,2)
decrypted = decrypt_affine(encrypted, inverse(1,26) ,2)

print("Original message: " + message)
print("Encrypted message: " + encrypted)
print("Decrypted message: " + decrypted)
```
